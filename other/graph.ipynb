```python
import csv

import matplotlib.pyplot as plt

epochs = []
train_loss = []
train_acc = []
test_acc = []

with open("results/resnet20_cifar_full_precision.csv", "r") as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        epochs.append(int(row["epoch"]))
        train_loss.append(float(row["train_loss"]))
        train_acc.append(float(row["train_acc"]))
        test_acc.append(float(row["test_acc"]))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(epochs, train_loss, linestyle="-", color="blue")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.set_title("Train Loss")
ax1.grid(True)

ax2.plot(epochs, train_acc, linestyle="-", color="green", label="Train")
ax2.plot(epochs, test_acc, linestyle="-", color="red", label="Test")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Accuracy")
ax2.set_title("Accuracy")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
